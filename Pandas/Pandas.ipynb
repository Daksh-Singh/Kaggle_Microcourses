```python
#to read a csv file we use read_csv method
reviews=pd.read_csv("winemag-data-130k-v2.csv")
#to check the first five rows of data we use head() 
reviews.head()
#similarly tail() method is use to check the last five rows of data
```

```python
#to get the mathematical information about data
reviews.describe()
```

```python
#to select a particlular column from reviews 
desc =reviews["description"]
#we can check the type of desc variable
type(desc)
```

```python
#Selecting the first value from description column
first_description = desc[0]
first_description
```

```python
#Selecting the first row from reviews data
first_row = reviews.loc[0]
first_row
```

```python
#Selecting the first 10 values from description column using loc
first_descriptions = reviews.description.loc[:9]
first_descriptions
#loc inlcude the last last index also where as iloc don't include the last index
```

```python
#Selecting the first 10 values from description column using iloc
first_descriptions = reviews.description.iloc[:10]
first_descriptions
```

```python
#to select recors with particular index labels we can use a list to pass in loc/iloc 
ind=[1,2,3,5,8]
sample_reviews = reviews.loc[ind]
sample_reviews.head()
```

```python
#to select paricular rows and columns from data we can again use lists
ind=[0,1,10,100]
cols=['country','province','region_1','region_2']
df = reviews.loc[ind,cols]
df
```

```python
#to create a dataframe of a particular value in columns
italian_wines = reviews.loc[reviews.country=='Italy']
italian_wines.head()
```

```python
#isin can be used to select multiple values of a column to create a dataframe along with various logical operators
top_oceania_wines = reviews.loc[(reviews.country.isin(['Australia','New Zealand'])) & (reviews.points>=95)]
top_oceania_wines.tail(1)
#will show only the last row of the data
```

```python
#various mathematical attributes of columns can also be calculated
median_points = reviews.points.median()
median_points
```

```python
#to check for the unique values in a column
countries = reviews.country.unique()
countries
```

```python
#value_counts will help us to count the duplicate values in a column
reviews_per_country = reviews.country.value_counts()
reviews_per_country.head()
```

```python
#we can update the columns values with various mathematical modification
#useful step in machine learning
centered_price = reviews['price']-reviews.price.mean()
centered_price.head()
```

```python
#with the help idmax() method we can return the index of particular row that has maximum value of selected columns
bargain_idx = (reviews.points / reviews.price).idxmax()
bargain_wine = reviews.loc[bargain_idx, 'title']
bargain_wine
```

```python
#we can use mapping and lambda function to retrive various values from the columns
#here we are counting number of time these words come
n_trop = reviews.description.map(lambda desc: "tropical" in desc).sum()
n_fruity = reviews.description.map(lambda desc: "fruity" in desc).sum()
descriptor_counts = pd.Series([n_trop, n_fruity], index=['tropical', 'fruity'])
descriptor_counts
```

```python
def stars(row):
    if row.country == 'Canada':
        return 3
    elif row.points >= 95:
        return 3
    elif row.points >= 85:
        return 2
    else:
        return 1

star_ratings = reviews.apply(stars, axis='columns')
#apply is an quivalent method to transform dataframe as map funtion do
star_ratings.head(1)
```

```python
#groupby is used to work on data using grps of various values
reviews_written = reviews.groupby(['taster_twitter_handle']).size()
reviews_written
```

```python
#various ways we can use groupby to extract the required data
best_rating_per_price = reviews.groupby('price')['points'].max()
best_rating_per_price.head()
```

```python
#agg takes the mathematical terms we want to calculate over a group of data
price_extremes = reviews.groupby(['variety']).price.agg([min,max])
#it will print min and max values of price of different variety of wines
price_extremes.head()
```

```python
#we can sort the values using various parameter by passing them in by attribute or sort_values
sorted_varieties = price_extremes.sort_values(by=['min', 'max'], ascending=False)
sorted_varieties.head()
```

```python
#we can use various mathematical functions along with groupby
reviewer_mean_ratings = reviews.groupby(['taster_name']).points.mean()
reviewer_mean_ratings.head()
```

```python
#Multi-index series 
country_variety_counts = reviews.groupby(['country', 'variety']).size().sort_values(ascending=False)
country_variety_counts.head()
```

```python
#dtype can be used to find the datatype of particular column values
dtype = reviews.points.dtype
print(dtype)
```

```python
#with the help o astype() we can modify the datatype of particular column values
point_strings = reviews['points'].astype('str')
```

```python
#isnull() function is used to check whether the value of that particular column is null or not
n_missing_prices = pd.isnull(reviews.price).sum()
n_missing_prices
```

```python
#fillna() is used to fill the empty values in the dataset,for eg we can fill with mean,'unknown',etc
reviews_per_region =reviews.region_1.fillna('Unknown').value_counts().sort_values(ascending=False)
reviews_per_region.head()
```

```python
#we can rename the columns of data using rename() function
renamed =reviews.rename(columns={'region_1':'region','region_2':'locale'})
renamed.head()
```

```python
#Setting the index name in dataset to wines
reindexed = reviews.rename_axis('wines',axis='rows')
reindexed.head(2)
```

```python
#renaming the rows in data
reindexed=reviews.rename(index={0: 'firstEntry', 1: 'secondEntry'})
reindexed.head(2)
```

```python
#Concat is used to combine different dataset so as to study them together
canadian_youtube = pd.read_csv("CAvideos.csv")
british_youtube = pd.read_csv("GBvideos.csv")

pd.concat([canadian_youtube, british_youtube])
```

```python
#Join is used to combine different DataFrame objects which have an index in common
left = canadian_youtube.set_index(['title', 'trending_date'])
right = british_youtube.set_index(['title', 'trending_date'])

left.join(right, lsuffix='_CAN', rsuffix='_UK')
```
