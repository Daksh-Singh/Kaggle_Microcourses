## Data Visualization Basics

```python
#Basic libraries required for the visualization
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
fifa_filepath = "fifa.csv"
#index_col will help us to denote first column with a different row
#parse_date helps the notebook to understand the each row label as a data
fifa_data = pd.read_csv(fifa_filepath, index_col="Date", parse_dates=True)
```

```python
fifa_data.head()    #to show the top five rows of dataset
```

```python
#to set the width and height of the plot respectively
plt.figure(figsize=(16,6))

# Line chart showing how FIFA rankings evolved over time
sns.lineplot(data=fifa_data)
#putting x and labels along with title
plt.xlabel("Year")
plt.ylabel("Ranking")
plt.title("FIFA rankings Over time")
```

```python
museum_filepath = "museum_visitors.csv"
museum_data = pd.read_csv(museum_filepath,index_col="Date",parse_dates=True)
```

```python
museum_data.tail()     #to review the last five rows of the data
```

```python
plt.figure(figsize=(16,6))
sns.set_style("ticks")
#This code will help us to use various themes or we can say varius seaborn styles
sns.lineplot(data=museum_data)
plt.xlabel("Year")
plt.ylabel("Visitors")
plt.title("Visitors in various Museums")
```

```python
# Line plot using a particular column of the data, here Avila Adobe visitors over years
plt.figure(figsize=(16,6))
sns.set_style("dark")
sns.lineplot(data=museum_data["Avila Adobe"])
plt.xlabel("Year")
plt.ylabel("Visitors")
plt.title("Visitors in Avila Adobe")
```

```python
ign_filepath = "ign_scores.csv"
ign_data = pd.read_csv(ign_filepath,index_col="Platform")
```

```python
#to review the data we will again use either head() or tail() function
ign_data.head()
```

```python
# Bar chart showing average score for racing games by platform
plt.figure(figsize=(16,6))
plt.title('Random shit')
sns.barplot(x=ign_data.index,y=ign_data["Racing"])
plt.ylabel("Avg scores")
plt.xlabel("Platforms")
# we cannot use ign_data["index_col"] here since we have initalized an index
```

```python
# Heatmap showing average game score by platform and genre
plt.figure(figsize=(16,6))
sns.heatmap(data=ign_data,annot=True)
#setting annot true will show the values in the respective blocks
plt.title("Heatmap")
plt.ylabel("Avg scores")
plt.xlabel("Platforms")
```

```python
candy_filepath = "candy.csv"
candy_data = pd.read_csv(candy_filepath,index_col="id")
```

```python
# to view the first five rows of the data
candy_data.head()
```

```python
#Scatter plot showing the relationship between 'sugarpercent' and 'winpercent'
plt.figure(figsize=(16,6))
sns.set_style("darkgrid")
sns.scatterplot(x=candy_data["sugarpercent"],y=candy_data["winpercent"])
plt.title("Scatter Plot")
plt.ylabel("Win percent")
plt.xlabel("Sugar percent")
```

```python
# Scatter plot with regression line showing the relationship between 'sugarpercent' and 'winpercent'
plt.figure(figsize=(16,6))
sns.set_style("white")
#By default style of seaborn plotting
sns.regplot(x=candy_data["sugarpercent"],y=candy_data["winpercent"])
plt.title("Scatter Plot with regression")
plt.ylabel("Win percent")
plt.xlabel("Sugar percent")
```

```python
# Scatter plot showing the relationship between 'pricepercent', 'winpercent', and 'chocolate' adding a third column with hue attribute
plt.figure(figsize=(16,6))
sns.set_style("whitegrid")
sns.scatterplot(x=candy_data["pricepercent"],y=candy_data["winpercent"],hue=candy_data["chocolate"]) 
plt.title("Scatter Plot with three columns")
plt.ylabel("Win percent")
plt.xlabel("price percent")
```

```python
# Color-coded scatter plot with regression lines
plt.figure(figsize=(16,6))
sns.set_style("darkgrid")
sns.lmplot(x="pricepercent",y="winpercent",hue="chocolate",data=candy_data) # Your code here
plt.title("Color coded Scatter Plot with regression line")
plt.ylabel("Win percent")
plt.xlabel("price percent")
```

```python
# Scatter plot showing the relationship between 'chocolate' and 'winpercent'
plt.figure(figsize=(16,6))
sns.set_style("darkgrid")
#The style i'll be using is darkgrid only
sns.swarmplot(x=candy_data["chocolate"],y=candy_data["winpercent"]) 
plt.title("Color coded Swarm Plot")
plt.ylabel("Choclate")
plt.xlabel("Win percent")
```

```python
cancer_b_filepath = "cancer_b.csv"
cancer_m_filepath = "cancer_m.csv"

cancer_b_data = pd.read_csv(cancer_b_filepath,index_col="Id")
cancer_m_data = pd.read_csv(cancer_m_filepath,index_col="Id")
```

```python
# to check the first five rows of the (benign) data
cancer_b_data.head()
```

```python
# Histograms for benign and maligant tumors
plt.figure(figsize=(16,6))
sns.set_style("darkgrid")
sns.distplot(a=cancer_b_data["Area (mean)"],kde=False) 
sns.distplot(a=cancer_m_data["Area (mean)"],kde=False)
#here make sure that there is no attribute of data name, we have to pass data to attribute 'a'
#Setting KDE true will change the shape of the bars making them curve, thus kde should be False
plt.title("Histogram for benign and maligant tumors mean area")
plt.ylabel("Number of observation")
plt.xlabel("Area(mean)")
```

```python
# Histograms for benign and maligant tumors, soft edges histogram
plt.figure(figsize=(16,6))
sns.set_style("darkgrid")
sns.distplot(a=cancer_b_data["Area (mean)"],kde=True) 
sns.distplot(a=cancer_m_data["Area (mean)"],kde=True)
plt.title("Histogram for benign and maligant tumors mean area")
plt.ylabel("Number of observation")
plt.xlabel("Area(mean)")
```

```python
# KDE plots for benign and malignant tumors
plt.figure(figsize=(16,6))
sns.set_style("darkgrid")
sns.kdeplot(data=cancer_b_data["Radius (worst)"],shade=True) 
sns.kdeplot(data=cancer_m_data["Radius (worst)"],shade=False)
#Setting shade True will shade the region under the graph and 
plt.title("Density plot for benign and maligant tumors worst radius")
plt.ylabel("Radius(worst)")
plt.xlabel("Number of obseravtion")
```

```python
#With the help of joinplot we can plot 2-D KDE plots
plt.figure(figsize=(16,6))
sns.jointplot(x=cancer_m_data['Area (mean)'], y=cancer_m_data['Perimeter (mean)'], kind="kde")
#kind will tell that we want kernel density estimate (KDE) plots
plt.title("2 D plot ")
plt.ylabel("Perimeter(mean)")
plt.xlabel("Area(mean)")
```
